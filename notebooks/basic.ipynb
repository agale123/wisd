```python
# Get events broken down by type
flat_events = sb.events(match_id=3795107, split=True, flatten_attrs=False)
flat_events.keys()
```

```python
passes = pd.DataFrame(flat_events['passes'])

def get_outcome(row):
    try:
        return row['outcome']['name']
    except Exception:
        return 'Complete'
passes['outcome'] = passes['pass'].apply(get_outcome)
```

```python
passes[passes['team'] == 'Belgium'].groupby(['outcome'])['outcome'].count()
```
